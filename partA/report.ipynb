## 2. Analysis of Friction Points

### A. The Decision Point between Try Sample / Skip for Later (58% Drop-off)

the user comes to the decision point page, only 42% of the users make it to the create assignment dashboard from here. In fact Users who click **“Try Sample”** (13.5% conversion) perform no better than users who click **“Skip”** (14.2%) when it comes to assignment creation. It's possible that this page is a little confusing, and it already presents a decision to them when theyre just trying the app which can be overwhelming. It looks like the "Sample" flow dumps users onto a dashboard full of mock data, which distracts them from the primary goal of creating the assignment.

### B. Create Assignment Dashboard to a successful creation (67% Drop-off)

Of the 1,728 users who actually start creating an assignment, two-thirds abandon it before finishing. The logs show users clicking "Add Topic" and "Add Problem," but dropping off. I suspect the editor requires too many decisions (topic / standard / problem / settings) for a first-time user who hasn't seen value yet, causing decision anxiety.


### C. Final Activation (70% Drop-off)

We see a massive drop between creating an assignment and a student joining ("Activation"). This step is a bit of a black box because we don't know if teachers are sharing the link and students are ignoring it, or if teachers simply aren't sharing it at all.

My hypothesis relies on an assumption: I assume the teacher is not sharing the link.
In practice, many teachers are cautious about introducing new technology in front of a live class, especially if they haven’t seen the student experience from end to end. If they are prepping alone and cannot see exactly what the student will see, they might not trust the tool enough to assign it on the next day. If the teacher is distributing the link and students aren't clicking, that would require a totally different diagnosis, but based on typical classroom behavior, that seems less likely.

## 3. Recommendations

### Better flow from Account Claiming to Assignment Creation

When a user selects their subject (e.g., *Algebra 1*) during signup, skip the dashboard entirely and drop them directly into the editor with a pre-populated assignment (maybe with common questions for a topic).


This helps with both creation drop-offs. It forces engagement (skipping the dashboard) and solves the decision point anxiety by giving them something that is ready to launch immediately. They just have to click "Assign."


Another idea is to use an interactive overlay (like distinct pulsing hotspots) to guide them: "We selected a topic for you. Click here to add a problem." This teaches the action required to activate.

Users can still explore other features later if needed, but this could ensure they reach value immediately during onboarding.

### Simulating Student View

Add a prominent **“Test as Student”** button on the final screen.

This addresses the assumption in Section 2C. By allowing teachers to simulate the student experience immediately (without a second device), we give them the confidence to share the link with a real class. It allows them to close the loop during their solo prep time.

### Google SSO  

We could add a **“Sign up with Google”** to bypass the phone number requirement. We lose a decent portion of traffic at phone verification, it's possible that a teacher isn't willing to go through the hassle of phone verification. A lot of teachers likely use Google Workspace, this reduces friction/privacy concerns while still verifying identity.
